```python
import pandas as pd
```

```python
data = pd.read_csv('../data/csv/deficit_lr_35->2.csv')
steps = [20, 40, 60, 80, 100, 120, 140][::-1]
```

```python
# write a function in pandas to extract from csv file columns which end with "epoch_acc/test"
def extract_acc(df):
    return df[[col for col in df.columns if col.endswith('epoch_acc/test')]]
extract_acc(data).shape
```

```python
#from every column in pandas dataframe extract 10 last non-nan value
def extract_last(df):
    return df.apply(lambda x: x.dropna().iloc[-5:])

values = extract_last(extract_acc(data)).median(axis=0).to_numpy()
```

```python
# plot the results in matplotlib given an array of values
def plot_results(steps, results, title):
    import matplotlib.pyplot as plt
    plt.plot(steps, results)
    plt.title(title)
    plt.show()
```

```python
plot_results(steps, values, 'deficit, test acc, lr: 0.35 -> 0.2, fpw: 0.0')
```
